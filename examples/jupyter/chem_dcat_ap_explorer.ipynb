# Chem-DCAT-AP Explorer notebook

This notebook is an interactive tool for exploring the [Chem-DCAT-AP](https://github.com/nfdi4chem/chem-dcat-ap) framework.
It illustrates how to use the Chem-DCAT-AP python data structures to create and manipulate metadata about chemistry related datasets.

This is useful, if one wants to create metadata for a dataset programmatically (e.g. by retrieving information from a database).

````python
from chem_dcat_ap.datamodel.chem_dcat_ap_pydantic import Catalogue, Dataset, Distribution, DataService
````

````python
chem_dataset = Dataset(
    title="Example Dataset",
    description="This is an example dataset for demonstration purposes.",
    issued="2024-01-01",
    modified="2024-01-02",
    identifier="https://example.com/dataset/1",
)
chem_catalogue = Catalogue(
    title="Example Catalogue",
    description="This is an example catalogue for demonstration purposes.",
    issued="2024-01-01",
    modified="2024-01-02",
    identifier="https://example.com/catalogue/1",
    dataset=[chem_dataset],
)
print(chem_catalogue.model_dump_json(indent=2))
````
